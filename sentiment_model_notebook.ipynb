```python
import pandas as pd
import spacy
```

```python
# Load the spaCy English language model
nlp = spacy.load("en_core_web_sm")

# Read the dataset from a CSV file
df = pd.read_csv('sentiment_analysis.csv')
df.head()
```

```python
df.info()
# Drop unwanted columns
selected_columns = ['name', 'reviews.text', 'reviews.clean', 'rating.sentiment']
df = df[selected_columns]
```

```python
# Check the distribution of labels 
df['rating.sentiment'].value_counts()
```

```python
df['vector'] = df['reviews.clean'].astype(str).apply(lambda text: nlp(text).vector) 
df.head()
```

```python
df = df.dropna(subset=['rating.sentiment'])
```

```python
from sklearn.model_selection import train_test_split

# Creating a train and test dataset
X_train, X_test, y_train, y_test = train_test_split(
    df['vector'].values,
    df['rating.sentiment'].values,
    test_size=0.2,
    random_state=2022
)
```
